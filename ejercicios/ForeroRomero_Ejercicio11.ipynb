# Ejercicio 3 (30 puntos)

```python
data = np.loadtxt('datos/lista_enteros.dat')

def distro(x, sigma, mu):
    return (1.0/(np.sqrt(2.0*np.pi)*sigma)) * (np.exp(-0.5 * (x-mu)**2/(sigma**2)))

valores_y, bins = np.histogram(data, bins=20, density=True)
valores_x = 0.5 * (bins[1:]+bins[0:-1])
plt.scatter(valores_x, valores_y, label='datos')

parameters, covariance = curve_fit(distro, valores_x, valores_y)
print(parameters)
x_model = np.linspace(valores_x.min(), valores_x.max(), 100)
y_model = distro(x_model, parameters[0], parameters[1])
plt.plot(x_model, y_model, label='modelo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

# 4. Ejercicio 4 (40 puntos)

```python
data = np.loadtxt('datos/lista_enteros.dat')

def distro(x, k):
    return (x**k) * np.exp(-x)/(scipy.special.factorial(k))

valores_y, bins = np.histogram(data, bins=20, density=True)
valores_x = 0.5 * (bins[1:]+bins[0:-1])
plt.scatter(valores_x, valores_y, label='datos')

parameters, covariance = curve_fit(distro, valores_x, valores_y)
print(parameters)
x_model = np.linspace(valores_x.min(), valores_x.max(), 100)
y_model = distro(x_model, parameters[0])
plt.plot(x_model, y_model, label='modelo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
