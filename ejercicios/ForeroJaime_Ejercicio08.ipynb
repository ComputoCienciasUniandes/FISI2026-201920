# 2. Scatter 3D

(2pts) Genere dos arrays s, t usando linspace desde cero hasta 2π

```python
s = np.linspace(0, 2.0*np.pi, 30)
t = np.linspace(0, 2.0*np.pi, 30)
```

(3pts) Con estos, genere un par de arrays x e y usando meshgrid como:

```python
x, y = np.meshgrid(s,t)
```

(5pts )Escriba una funci ́on $f(x,y) = e^{−x/\pi} \cos x \sin y$. Con la funcion genere un array z = f (x, y)

```python
def f(x, y):
    return np.exp(-x/np.pi) * np.cos(x) * np.sin(y)
z = f(x,y)
```

(5 pts) Imprima la forma (shape) de los arrays s,t,x,y,z. Explique en
comentarios que hace la funcion meshgrid.

```python
print(np.shape(s))
print(np.shape(t))
print(np.shape(x))
print(np.shape(y))
print(np.shape(z))
# La funcion meshgrid crea un array bidimensional a partir de arrays unidimensional
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection="3d")
ax.scatter(x, y, z)
ax.set_xlabel("X Label") 
ax.set_ylabel("Y Label")
ax.set_zlabel("Z Label")
```
